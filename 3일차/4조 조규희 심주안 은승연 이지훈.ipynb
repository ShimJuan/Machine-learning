- 남성, 53 달러인 경우 생존 유무를 예측하시오
- w, b를 구하고, 정확도를 구하시오 (텐서, 케라스)

```python
df = titanic[['sex','fare','survived']]
# df['winning_rate'] = df.apply(lambda x:relative_record(x['team'], x['against'])[3], axis=1)
```

```python
def sex(x):
    if x == 'male':
        return(1)
    else:
        return(0)
```

```python
df['sex'] = df.apply(lambda x: sex(x['sex']), axis=1)
```

```python
# 정규화

from sklearn import preprocessing

scale = preprocessing.MinMaxScaler()
dataN = scale.fit_transform(df)
```

```python
df
```

```python
xdata = dataN[:,:-1]
ydata = dataN[:,[-1]]
```

```python
x = tf.placeholder(tf.float32, shape=[None,2])
y = tf.constant(ydata, tf.float32)
```

```python
w = tf.Variable(tf.random_uniform([2,1]))
b = tf.Variable(tf.random_uniform([1]))
```

```python
z = tf.matmul(x,w) + b
hx = tf.nn.sigmoid(z)
```

```python
cost_i = y*(-tf.log(hx)) + (1-y)*(-tf.log(1-hx))
cost = tf.reduce_mean(cost_i)
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.001)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
predict = tf.cast(hx>0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predict, y), dtype = tf.float32))
```

```python
for i in range(100000):
    sess.run(train, {x:xdata})
    if i % 1000 == 0:
        print(sess.run(accuracy, {x:xdata}))
```
